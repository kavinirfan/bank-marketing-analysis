# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first few rows
df.head()
```

```python

# Check for missing values and data types
print("Data types and non-null counts:\n")
df.info()

print("\nMissing values per column:\n")
print(df.isnull().sum())
```

```python

# Summary statistics
df.describe()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable
plt.figure(figsize=(6,4))
ax = sns.countplot(x='y', data=df, hue='y', palette=['#4C72B0','#DD8452'], legend=False)
plt.title('Distribution of Target Variable (y): Term Deposit Subscription')
plt.xlabel('Subscribed to Term Deposit')
plt.ylabel('Number of Clients')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x()+p.get_width()/2, p.get_height()), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python

# Correlation matrix for numerical features
num_df = df.select_dtypes(include='number')
corr = num_df.corr()

plt.figure(figsize=(9,7))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', square=True, linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
```


## Conclusion

In this analysis, we explored the structure and composition of the bank marketing dataset. Key takeaways include:

- The dataset contains **41,188 records** and **21 columns**, with **no missing values** in any column — the data is clean and ready for further processing.
- The dataset includes customer demographics (age, job, marital status, education), financial details (default, housing, loan), and campaign contact details, which can be useful for segmentation.
- The target variable `y` is **imbalanced**: only about **11.3% (4,640) of clients subscribed** to a term deposit, while **88.7% (36,548) did not**. This imbalance should be addressed (e.g. via resampling or class weighting) before building a predictive model.
- The correlation matrix shows that several macroeconomic indicators (`emp.var.rate`, `euribor3m`, `nr.employed`, `cons.price.idx`) are **strongly correlated with each other**, suggesting potential multicollinearity that should be handled during feature selection.
- `pdays` and `previous` also show a notable relationship, reflecting that clients contacted more recently in past campaigns tend to have lower `pdays` values.
- Further preprocessing, feature engineering, and modeling would be the next steps for predictive analysis.

This initial inspection sets the foundation for a deeper analysis or machine learning workflow to improve marketing effectiveness.
